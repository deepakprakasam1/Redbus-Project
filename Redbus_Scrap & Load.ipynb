```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.action_chains import ActionChains
import pyautogui as py

upsrtc = []



def get_bus_details(i, current_page_index):
    time.sleep(10)

    wait = WebDriverWait(driver, 100)
    buttons = driver.find_elements(By.XPATH, '//i[contains(@class, "p-left-10 icon icon-down")]')
    data_found = False

    for button in buttons:
        try:
            driver.execute_script("arguments[0].scrollIntoView(true);", button)
            time.sleep(1)
            driver.execute_script("arguments[0].click();", button)
            time.sleep(3)
        except Exception as e:
            print(f"An error occurred while clicking a button: {e}")
            continue

    wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".clearfix.bus-item-details")))

    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    buses = driver.find_elements(By.XPATH, '//div[@class="clearfix bus-item-details"]')

    if not buses:
        try:
            specific_container = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CLASS_NAME, "sodc-doj-c-date-time")))
            specific_container.click()
            time.sleep(5)
            wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".clearfix.bus-item-details")))
            buses = driver.find_elements(By.XPATH, '//div[@class="clearfix bus-item-details"]')
        except Exception as e:
            print(f"An error occurred while trying to get bus details for upcoming days: {e}")
    
    for bus in buses:
        name = bus.find_element(By.CSS_SELECTOR, ".travels").text
        type = bus.find_element(By.CSS_SELECTOR, ".bus-type").text
        departure_time = bus.find_element(By.CSS_SELECTOR, ".dp-time").text
        arrival_time = bus.find_element(By.CSS_SELECTOR, ".bp-time").text
        duration = bus.find_element(By.CSS_SELECTOR, ".dur").text
        price = bus.find_element(By.CSS_SELECTOR, ".fare").text
        seats_available = bus.find_element(By.CSS_SELECTOR, ".seat-left").text
        try:
            rating = bus.find_element(By.CSS_SELECTOR, ".rating-sec .rating").text
        except:
            rating = "No rating"

        upsrtc.append({
            "route": i,
            "name": name,
            "type": type,
            "departure_time": departure_time,
            "arrival_time": arrival_time,
            "duration": duration,
            "price": price,
            "seats_available": seats_available,
            "rating": rating,
        })

        data_found = True

    return data_found
   

def navigation(current_page_index):
    max_attempts = 10
    attempts = 0
    while attempts < max_attempts:
        py.hotkey("alt", "left")
        time.sleep(1)
        try:
            WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, ".DC_117_pageTabs"))
            )
            break
        except:
            attempts += 1

    # Ensure we are back on the correct page
    pages = driver.find_elements(By.CSS_SELECTOR, ".DC_117_pageTabs")
    page = pages[current_page_index]
    action.move_to_element(page).click().perform()
    time.sleep(3)  # Adjust sleep time if necessary
    return data_found


driver = webdriver.Chrome()
driver.get("https://www.redbus.in/online-booking/uttar-pradesh-state-road-transport-corporation-upsrtc/")
action = ActionChains(driver)

pages = driver.find_elements(By.CSS_SELECTOR, ".DC_117_pageTabs")
for page_number in range(len(pages)):
    pages = driver.find_elements(
        By.CSS_SELECTOR, ".DC_117_pageTabs"
    )  # Refresh the page elements list
    page = pages[page_number]

    try:
        action.move_to_element(page).click().perform()
        time.sleep(3)
        route_collected = []
        routes = driver.find_elements(By.CLASS_NAME, "route")
        for route in routes:
            route_name = route.text
            if route_name not in route_collected:
                route_collected.append(route_name)

        for route_name in route_collected:
            try:
                route_element = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.LINK_TEXT, route_name))
                )
                action.move_to_element(route_element).click().perform()
                data_found = get_bus_details(route_name, page_number)
                navigation(page_number)
                
                
              
            except Exception as e:
                print(f"An error occurred while processing route '{route_name}': {e}")
                navigation(page_number)
    except Exception as e:
        print(f"An error occurred while navigating pages: {e}")

driver.quit()
```
